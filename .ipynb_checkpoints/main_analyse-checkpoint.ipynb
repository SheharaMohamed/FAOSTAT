#### Calculer pour chaque pays et pour chaque produit, la disponibilité alimentaire en Kcal et en Kg de proétines.

```python
data = sql_to_df('''SELECT pays, annee, produit, supply.supply AS 'Food supply (Kcal/yr)', 
supply_quantity.supply_quantity AS 'Food supply quantity (Kg/yr)', prot.Prot AS 'Proteines (Kg/yr)'
FROM (
	SELECT pays, produit, annee, ROUND(valeur*population*365,1) AS 'supply'
	FROM consomme C, produit P, pays, bilan B, population, annee A
	WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage 
    AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
	AND usages = 'Food supply (kcal/capita/day)'
	GROUP BY pays,produit, annee) supply
INNER JOIN
    (SELECT pays, produit, annee, ROUND(valeur*population*365/1000,1) AS 'Prot'
	FROM consomme C, produit P, pays, bilan B, population , annee A
	WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
    AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
	AND usages = 'Protein supply quantity (g/capita/day)'
	GROUP BY pays,produit, annee) prot using (pays, produit, annee)
INNER JOIN
	(SELECT pays, produit, annee, ROUND(valeur*population,1) AS 'supply_quantity'
	FROM consomme C, produit P, pays, bilan B, population , annee A
	WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
    AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
	AND usages = 'Food supply quantity (kg/capita/yr)'
	GROUP BY pays,produit, annee) supply_quantity using(pays,produit,annee)''')
```

#### Calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal ?

```python
df = data.copy()
#df = df[df['produit']!= 'Cocoa Beans and products']

#Ignorons les valeurs où la dispo alimentaire est égale à 0.
df = df[df['Food supply quantity (Kg/yr)'] > 0]

df['ratio (Kcal/Kg)'] = df['Food supply (Kcal/yr)'] / df['Food supply quantity (Kg/yr)']
df['% Proteine'] = df['Proteines (Kg/yr)'] / df['Food supply quantity (Kg/yr)']*100

df_prod = round(df.groupby(['produit','annee'], as_index=False).mean(),2)

round(df_prod['ratio (Kcal/Kg)'][df_prod.produit=='Eggs'].mean())
```

#### Calculer, pour les produits végétaux uniquement, la disponibilité intérieure mondiale exprimée en kcal et en Kg protéines pour chaque année et tracer la viz correspondante ?

Pour calculer la disponibilité intérieure mondiale, nous devons récupérer dans notre base de données les valeurs de 'Domestic supply quantity' pour chaque produit, pays et année.

```python
data1 = sql_to_df('''SELECT produit, id_cat, annee, (valeur*1000*1000) AS 'Domestic supply quantity (kg/yr)'
	FROM consomme C, produit P, pays, bilan B, population , annee A
	WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
    AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
	AND usages = 'Domestic supply quantity'
	GROUP BY produit, annee, pays ;''')
```

```python
df_domestic = data1.copy()

#On enleve les produits animaux du dataframe
df_domestic = df_domestic[df_domestic['id_cat']!= 1]
del df_domestic['id_cat']

#On groupe les résultats par produit et année et on additionne les domestics supply quantity
df_domestic = df_domestic.groupby(['produit', 'annee'], as_index=False).sum()

#Finalement, on merge notre dataframe avec celui des produits pour récupérer le ratio
df_domestic = df_domestic.merge(df_prod[['produit','annee','ratio (Kcal/Kg)','% Proteine', 'Proteines (Kg/yr)']], on=['produit','annee'])

#On calcule la disponibilité pour chaque produit végétal, pour chaque année
df_domestic['Disponibilité Végétaux (kcal)'] = df_domestic['Domestic supply quantity (kg/yr)'] * df_domestic['ratio (Kcal/Kg)']

df_domestic
```

```python
graph = df_domestic.groupby('annee')['Disponibilité Végétaux (kcal)'].sum()

prot = df_domestic.copy()
prot['Domestic protein supply (kg/yr)']= prot['Domestic supply quantity (kg/yr)']*(prot['% Proteine']/100)
prot = prot.groupby('annee')['Domestic protein supply (kg/yr)'].sum()

sns.set_style("white")

g, ax = plt.subplots(figsize=(12, 5))
g = sns.lineplot(data=graph, marker = 'o')

ax2 = ax.twinx() #twin x axis
sns.lineplot(data=prot, marker = 'o', color="darkred")

sns.despine(trim=True, right=False, offset=2)

g.set_xticks(range(2014,2019,1))

ax.set_xlabel(None)
ax.set_ylabel("Disponibilité Végétaux (kcal)", size=12)
ax.set_title("Disponibilité intérieure mondiale des végétaux par an", size=18)

ax2.set_ylabel("Disponibilité des protéines(kg)",size=15)
ax2.yaxis.label.set_color('darkred')
ax2.spines["right"].set_edgecolor('darkred')
ax2.tick_params(axis='y', colors ='darkred')
```

#### Combien d'humains pourraient être nourris si toute la disponibilité intérieure mondiale de produits végétaux était utilisée pour de la nourriture ?

Donnez les résultats en termes de calories, puis de protéines, et exprimez ensuite ces 2 résultats en pourcentage de la population mondiale.

Nous avons calculés précédemment la disponibilité pour chaque année. Pour savoir combien d'humaines pourraient être nourris à partir de cette disponibilité, nous allons considérer un apport de référence de 2100 Kcal/jour et 60g/j de protéines par personne (source : https://www.who.int/nutrition/publications/en/nut_needs_emergencies_text.pdf).

```python
df_population = sql_to_df('''SELECT annee, SUM(population) as 'population'
FROM population P, annee A
WHERE P.code_annee = A.code_annee
GROUP BY P.code_annee;''')

#On merge un df contenant la population par année avec la disponibilité en végétaux par an
df_dispo = df_population.merge(df_domestic.groupby('annee')['Disponibilité Végétaux (kcal)'].sum(), on ='annee')

df_dispo
```

```python
kcal_year = 365 * 2100
df_dispo['%kcal disponible par habitant'] = round((df_dispo['Disponibilité Végétaux (kcal)']/kcal_year/ df_dispo['population'])*100)
df_dispo
```

En prenant un apport de référence de 2000 Kcal/j par personne, on constate qu'on aurait pu nourrir 2,5 fois la population mondiale en 2018 ! Avec un apport de référence un peu plus élevé (2700 kcal/j), nos résultats tournent plutôt aux alentours de 200%.

```python
df_prot = df_domestic.copy()
df_prot['Domestic protein supply (kg/yr)']= df_prot['Domestic supply quantity (kg/yr)']*(df_prot['% Proteine']/100)
df_prot = df_prot.groupby('annee', as_index=False)['Domestic protein supply (kg/yr)'].sum()
df_prot
```

```python
df_dispo = df_dispo.merge(df_prot, on ='annee')

proteine_year = (60*365)/1000
df_dispo['%proteine disponible par habitant'] = round((df_dispo['Domestic protein supply (kg/yr)']/proteine_year/ df_dispo['population'])*100)
df_dispo
```

On retrouve des résultats sensiblement équivalent à ceux calculés précédemment. Avec la disponibilité intérieure mondiales de végétaux en protéine, on pourrait théoriquement nourrir plus de deux fois la population mondiale. Et cela tend à augmenter avec les années !

#### Combien d'humains pourraient être nourris si toute la disponibilité alimentaire en produits végétaux (Food), la nourriture végétale destinée aux animaux (Feed) et les pertes de produits végétaux (Waste) étaient utilisés pour de la nourriture ? 

Donnez les résultats en termes de calories, puis de protéines, et exprimez ensuite ces 2 résultats en pourcentage de la population mondiale.

```python
food = sql_to_df('''SELECT produit, id_cat,pays, annee, (valeur*1000*1000) AS 'Valeur'
		FROM consomme C, produit P, pays, bilan B, population , annee A
		WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
		AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
		AND usages = 'Food' AND id_cat <> 1
		GROUP BY produit, annee, pays;
''')

feed = sql_to_df('''SELECT produit, id_cat,pays, annee, (valeur*1000*1000) AS 'Valeur'
		FROM consomme C, produit P, pays, bilan B, population , annee A
		WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
		AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
		AND usages = 'Feed' AND id_cat <> 1
		GROUP BY produit, annee, pays;
''')

losses = sql_to_df('''SELECT produit, id_cat,pays, annee, (valeur*1000*1000) AS 'Valeur'
		FROM consomme C, produit P, pays, bilan B, population , annee A
		WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
		AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
		AND usages = 'Losses' AND id_cat <> 1
		GROUP BY produit, annee, pays;
''')
```

```python
df_food = food.copy()
df_food = df_food.groupby(['produit', 'annee'], as_index=False).sum()
df_food = df_food.merge(df_prod[['produit','annee','ratio (Kcal/Kg)','% Proteine']], on=['produit','annee'])
df_food['Total Dispo (kcal)'] = df_food['Valeur'] * df_food['ratio (Kcal/Kg)']
df_food['Total Dispo (kg de Proteine)']= df_food['Valeur']*(df_food['% Proteine']/100)
df_food = df_food.groupby('annee')[['Total Dispo (kcal)','Total Dispo (kg de Proteine)']].sum()

df_feed = feed.copy()
df_feed = df_feed.groupby(['produit', 'annee'], as_index=False).sum()
df_feed = df_feed.merge(df_prod[['produit','annee','ratio (Kcal/Kg)','% Proteine']], on=['produit','annee'])
df_feed['Total Dispo (kcal)'] = df_feed['Valeur'] * df_feed['ratio (Kcal/Kg)']
df_feed['Total Dispo (kg de Proteine)']= df_feed['Valeur']*(df_feed['% Proteine']/100)
df_feed = df_feed.groupby('annee')[['Total Dispo (kcal)','Total Dispo (kg de Proteine)']].sum()

df_losses = losses.copy()
df_losses = df_losses.groupby(['produit', 'annee'], as_index=False).sum()
df_losses = df_losses.merge(df_prod[['produit','annee','ratio (Kcal/Kg)','% Proteine']], on=['produit','annee'])
df_losses['Total Dispo (kcal)'] = df_losses['Valeur'] * df_losses['ratio (Kcal/Kg)']
df_losses['Total Dispo (kg de Proteine)']= df_losses['Valeur']*(df_losses['% Proteine']/100)
df_losses = df_losses.groupby('annee')[['Total Dispo (kcal)','Total Dispo (kg de Proteine)']].sum()
```

```python
df_dispo_tot = pd.concat([df_feed, df_food,df_losses],axis=0)
# df = df.groupby('col1').sum().reset_index()

df_dispo_tot = df_dispo_tot.sort_index()
df_dispo_tot = df_dispo_tot.groupby('annee').sum()
df_dispo_tot = df_dispo_tot.merge(df_population.groupby('annee').sum(), on ='annee')

df_dispo_tot['Disponibilité % (kcal)'] = round((df_dispo_tot['Total Dispo (kcal)']/kcal_year/ df_dispo_tot['population'])*100)
df_dispo_tot['Disponibilité % (kg de protéine)'] = round((df_dispo_tot['Total Dispo (kg de Proteine)']/proteine_year/ df_dispo_tot['population']*100))

df_dispo_tot
```

#### Combien d'humains pourraient être nourris avec la disponibilité alimentaire mondiale ? 
Donnez les résultats en termes de calories, puis de protéines, et exprimez ensuite ces 2 résultats en pourcentage de la population mondiale.

```python
data_monde = sql_to_df('''SELECT produit, annee, (valeur*1000*1000) AS 'Domestic supply quantity (kg/yr)'
	FROM consomme C, produit P, pays, bilan B, population , annee A
	WHERE C.id_pays = pays.id_pays AND C.code_produit = P.code_produit AND B.code_usage=C.code_usage
    AND population.id_pays = pays.id_pays AND A.code_annee = C.code_annee
	AND usages = 'Domestic supply quantity'
	GROUP BY produit, annee, pays ;''')
```

```python
df_monde = data_monde.copy()

#On groupe les résultats par produit et année et on additionne les domestics supply quantity
df_monde = df_monde.groupby(['produit', 'annee'], as_index=False).sum()

#Finalement, on merge notre dataframe avec celui des produits pour récupérer le ratio
df_monde = df_monde.merge(df_prod[['produit','annee','ratio (Kcal/Kg)','% Proteine', 'Proteines (Kg/yr)']], on=['produit','annee'])

#On calcule la disponibilité pour chaque produit, pour chaque année
df_monde['Disponibilité Mondiale (kcal)'] = df_monde['Domestic supply quantity (kg/yr)'] * df_monde['ratio (Kcal/Kg)']
df_monde['Disponibilité Mondiale Proteine (kg/yr)']= df_monde['Domestic supply quantity (kg/yr)']*(df_monde['% Proteine']/100)

#On groupe les résultats par année
df_monde = df_monde.groupby('annee')[['Disponibilité Mondiale (kcal)','Disponibilité Mondiale Proteine (kg/yr)']].sum()

#On merge avec le dataframe population
df_monde = df_monde.merge(df_population, on ='annee')


#On exprime les résultats en % de la population mondiale 
df_monde['%kcal disponible par habitant'] = round((df_monde['Disponibilité Mondiale (kcal)']/kcal_year/ df_monde['population'])*100)
df_monde['%proteine disponible par habitant'] = round((df_monde['Disponibilité Mondiale Proteine (kg/yr)']/proteine_year/ df_monde['population'])*100)

df_monde
```

Si toute la disponibilité intérieure mondiale était destinée à nourrir les humains, on pourrait théoriquement nourrir plus de trois fois la population mondiale (aussi bien en terme de Kcal que de Kg de protéines) !

### Parmi les données des bilans alimentaires au niveau mondial, sélectionner les 200 plus grandes importations de ces produits (1 importation = une quantité d'un produit donné importée par un pays donné sur l'année choisie)

```python
importation = bilan[(bilan['usages'] == 'Import Quantity') & bilan['produit'].isin(exportation.index)][['pays','produit','valeur','usages']].sort_values(by='valeur', ascending=False)[:200]
importation.head()
```

### Grouper ces importations par produit, et calculer pour chaque produit :
* Le ratio entre la quantité destinés aux "Autres utilisations" (Other uses) et la disponibilité intérieure.
* Le ratio entre la quantité destinée à la nourriture animale et la quantité destinée à la nourriture (animale + humaine)
* Top 3 des produits qui ont la plus grande valeur pour chacun des 2 ratios

```python
usages_list = ['Import Quantity', 'Domestic supply quantity','Feed','Other uses (non-food)','Food']
```

```python
product_import = bilan[(bilan['usages'].isin(usages_list)) & bilan['produit'].isin(exportation.index)][['pays','produit','valeur','usages']]
product_import = product_import.groupby(['produit','usages'], as_index=False).sum()

product_import.sample(3)
```

```python
product_import_pivot = pd.pivot(data = product_import, index = 'produit', columns='usages', values='valeur').reset_index().rename_axis(None, axis=1)
product_import_pivot.fillna(0, inplace=True)
product_import_pivot
```

```python
product_import_pivot['Other uses/Domestic supply (%)'] = round(product_import_pivot['Other uses (non-food)']/product_import_pivot['Domestic supply quantity']*100,2)
product_import_pivot['Feed/Food (%)'] = round(product_import_pivot['Feed']/(product_import_pivot['Food']+product_import_pivot['Feed'])*100,2)
product_import_pivot
```
