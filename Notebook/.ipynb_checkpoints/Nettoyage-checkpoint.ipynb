```python
import pandas as pd
import numpy as np
```

## Produits

```python
df_2018 = pd.read_csv('../data/FAOSTAT_data_vegetable_pro_2018.csv')
df_2017 = pd.read_csv('../data/FAOSTAT_data_vegetable_pro_2017.csv')
df_2016 = pd.read_csv('../data/FAOSTAT_data_vegetable_pro_2016.csv')
df_2015 = pd.read_csv('../data/FAOSTAT_data_vegetable_pro_2015.csv')
df_2014 = pd.read_csv('../data/FAOSTAT_data_vegetable_pro_2014.csv')
df_produits_vegetal = pd.concat([df_2018,df_2017,df_2016,df_2015,df_2014])
df_produits_vegetal.head()
```

```python
df_produits_vegetal.info()
```

```python
df_produits_animal = pd.read_csv('../data/FAOSTAT_data_animal_pro.csv')
df_produits_animal.head()
```

```python
df_produits_animal.info()
```

## Population

```python
df_population = pd.read_csv('../data/FAOSTAT_data_Populations.csv')
df_population.head()
```

```python
df_population.info()
```

```python
df_population.dropna(axis=1,how='any',inplace=True)
df_population.head()
```

## Sousnutrition

```python
df_sousnutrition = pd.read_csv('../data/FAOSTAT_data_Undernourished.csv')
df_sousnutrition.head()
```

```python
df_sousnutrition.info()
```

```python
df_sousnutrition.dropna(axis=1,how='all',inplace=True)
df_sousnutrition.dropna(axis=0,inplace=True)
df_sousnutrition.head()
```

## Cereal

```python
df_cereal = pd.read_csv('../data/FAOSTAT_data_Cereal.csv')
df_cereal.head()
```

```python
df_cereal.info()
```

## Seperation des tables pour tlcharger  base de donnes SQL

```python
#Category
df_category = pd.DataFrame(data={'Category_code':[1,2,3],'Category':['Vegetal product','Animal product','Vegetal Product'],'Sub Category':['General','General','Cereal']})
df_category.to_csv('../sql_upload/category.csv',index=False,sep='\t')
```

```python
#Produits
df_produits_vegetal.loc[:,'Category_code'] = 1
df_produits_animal.loc[:,'Category_code'] = 2

df_final_produit = pd.concat([df_produits_vegetal,df_produits_animal],axis=0)
df_produit = df_final_produit[['Item Code','Item','Category_code']].drop_duplicates()

df_temp = df_cereal[['Item Code','Item']].drop_duplicates()
df_produit.Category_code[df_produit.Item.isin(df_temp.Item)] = 3
df_produit['Item'] = '"'+df_produit['Item']+'"'

df_produit.head(70)

df_produit.to_csv('../sql_upload/produits.csv',index=False,sep='\t')
```

```python
#Information
df_final_produit[['Element Code','Element','Unit']].drop_duplicates().sort_values(by='Element Code').to_csv('../sql_upload/information.csv',index=False,sep='\t')
```

```python
#Country
cols = ['Area Code','Area']
df_country = pd.concat([df_produits_vegetal[cols],df_produits_animal[cols],df_cereal[cols],df_population[cols],df_sousnutrition[cols]],axis=0)
df_country.drop_duplicates().sort_values(by='Area Code').to_csv('../sql_upload/country.csv',index=False,sep='\t')
```

```python
#Year
cols = ['Year Code','Year']
df_year = pd.concat([df_produits_vegetal[cols],df_produits_animal[cols],df_cereal[cols],df_population[cols],df_sousnutrition[cols]],axis=0)
df_year.drop_duplicates().sort_values(by='Year Code').to_csv('../sql_upload/year.csv',index=False,sep='\t')
```

```python
#product_infromation
df_final_produit[['Area Code','Year Code','Item Code','Element Code','Value']].drop_duplicates().to_csv('../sql_upload/product_infromation.csv',index=False,sep='\t')
```

```python
# df_sousnutrition['Years considered']

def year_trans(yr):
    res = []
    fir = int(yr.split('-')[0])
    sec = int(yr.split('-')[1])
    
    for i in range(fir,sec+1):
        res.append(i)
    return res

df_sousnutrition['Years considered'] = df_sousnutrition['Year'].apply(year_trans)
df_sousnutrition['Year Code'] = df_sousnutrition['Year'].apply(lambda x: year_trans(x)[1])
df_sousnutrition['Year'] = df_sousnutrition['Year'].apply(lambda x: year_trans(x)[1])

df_sousnutrition['Years considered'] = '"'+df_sousnutrition['Years considered'].apply(lambda x: ','.join(str(i) for i in x))+'"'
df_sousnutrition[['Year','Year Code','Years considered']]
```

```python
df_population[df_population['Area'].str.contains('China,', case=False)]
```

```python
#population
df_population[['Area Code','Year Code','Unit','Value']].drop_duplicates().to_csv('../sql_upload/population.csv',index=False,sep='\t')

#undernourished
df_sousnutrition['Value'] = df_sousnutrition['Value'].apply(lambda x: float(x.split('<')[1]) if len(x.split('<'))>1 else float(x.split('<')[0]))
df_sousnutrition[['Area Code','Year Code','Years considered','Unit','Value']].drop_duplicates().to_csv('../sql_upload/undernourished.csv',index=False,sep='\t')
```
